```python
import torch
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
%matplotlib inline


def scale_plot_size(factor=1.5):
    import matplotlib as mpl
    default_dpi = mpl.rcParamsDefault['figure.dpi']
    mpl.rcParams['figure.dpi'] = default_dpi*factor


def load_data():
    m = torch.load("/data1/yangming/.cache/torch/hub/checkpoints/detr-r50-e632da11.pth", map_location="cpu")["model"]
    query_embed = m["query_embed.weight"]
    
    return query_embed


def vis():
    x = load_data().numpy()
    
    tsne = TSNE(n_components=2, metric="cosine")
    x = tsne.fit_transform(x)
    print(x.shape)
    
    scale_plot_size()
    plt.scatter(x[..., 0], x[..., 1])
    plt.show()
    

vis()
```
